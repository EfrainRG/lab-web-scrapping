#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page

````python
# This is the url you will scrape in this exercise
import re
urlarb ='https://es.wikipedia.org/wiki/Historia_de_Python' 
htmlarb = requests.get(urlarb).content
souparb= BeautifulSoup(htmlarb)
````

````python
pylinks = souparb.find_all('a')
#pylinks[0]
pyurl = []
for i in pylinks: 
    obtener = str(i.get('href'))
    if obtener.startswith('/'):
        # como directamente encontramos el link, pero no está añadida protocólo
        # lo añadimos como como una suma de cadenas
        put = 'https://es.wikipedia.org' + obtener
        pyurl.append(put)
    elif obtener.startswith('http'):
        # esta sentencia fue añadida porque hay varios elementos encontrados en
        # 'pylinks' que ya son url directamente, así que solo serán añadidas
        # a la lista de urls
        pyurl.append(i)
        
pyurl
````
